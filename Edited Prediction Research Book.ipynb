```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings("ignore")
```

```python
gray_img = cv2.imread('images/noedit.jpg', cv2.IMREAD_GRAYSCALE)
imgplot = plt.imshow(gray_img)
```

```python
hist = cv2.calcHist([gray_img],[0],None,[256],[0,256])
plt.hist(gray_img.ravel(),256,[0,256])
# plt.hist(hist,256,[0,256])
plt.title('Histogram for gray scale picture')
plt.show()
```

```python
# All histograms 

f, axes = plt.subplots(4, 2, figsize=(12,8), sharex=True, sharey=False)
gray_img = cv2.imread('images/noedit.jpg', cv2.IMREAD_GRAYSCALE)
p1 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[0,0])
axes[0,0].set_title('No edit')

gray_img = cv2.imread('images/low_low.jpg', cv2.IMREAD_GRAYSCALE)
p2 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[1,0])
axes[1,0].set_title('Low edit -low')
gray_img = cv2.imread('images/low_high.jpg', cv2.IMREAD_GRAYSCALE)
p3 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[1,1])
axes[1,1].set_title('Low edit -high')

gray_img = cv2.imread('images/mid_low.jpg', cv2.IMREAD_GRAYSCALE)
p4 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[2,0])
axes[2,0].set_title('Mid edit -low')
gray_img = cv2.imread('images/mid_high.jpg', cv2.IMREAD_GRAYSCALE)
p5 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[2,1])
axes[2,1].set_title('Mid edit -high')

gray_img = cv2.imread('images/high_low.jpg', cv2.IMREAD_GRAYSCALE)
p6 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[3,0])
axes[3,0].set_title('High edit -low')
gray_img = cv2.imread('images/high_high.jpg', cv2.IMREAD_GRAYSCALE)
p7 = sns.distplot(gray_img.ravel(), bins=256, ax=axes[3,1])
axes[3,1].set_title('High edit -high')


p1.set(xlim=(0,256))
```
